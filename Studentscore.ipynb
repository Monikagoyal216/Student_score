```python
# Separate the features (study hours) and target variable (score)
X = data['Hours'].values.reshape(-1, 1)
y = data['Scores'].values

# Create and train a linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict the score for 9.5 hours of study
hours = 9.5
predicted_score = model.predict([[hours]])

print("Predicted score for studying 9.5 hours:", predicted_score[0])
```

```python
plt.scatter(X, y, color='blue', label='Actual Scores')
plt.plot(X, model.predict(X), color='red', label='Linear Regression')
plt.scatter(hours, predicted_score, color='green', label='Predicted Score')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.title('Linear Regression - Hours Studied vs. Scores')
plt.legend()
plt.show()
```
